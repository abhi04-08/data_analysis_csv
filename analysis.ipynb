```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load CSV file
df = pd.read_csv("sales.csv")

# Step 3: Display few rows
print("First 5 datasets in a row:")
display(df.head())

# Step 4: Check the basic info
print("\nDataset Info:")
display(df.info())

# Step 5: Check the basic stats
print("\nDataset Statistics:")
display(df.describe())

# Step 6: Group by Product
sales_by_product = df.groupby("Product")["Sales"].sum()
print("\nSales by product")
print(sales_by_product)

# Step 7: Group by Region
sales_by_region = df.groupby("Region")["Sales"].sum()
print("\nSales by Region")
print(sales_by_region)

# Step 8: Plot sales by product
plt.figure(figsize=(6,4))
sales_by_product.plot(kind="bar", color="skyblue")
plt.title("Total Sales by Product")
plt.ylabel("Sales")
plt.show()

# Step 9: Plot sales by region
plt.figure(figsize=(6,4))
sales_by_region.plot(kind="bar", color="lightgreen")
plt.title("Total Sales by Region")
plt.ylabel("Sales")
plt.show()

# Step 10: Time Series Plot
plt.figure(figsize=(8,4))
df.groupby("Date")["Sales"].sum().plot(kind="line", marker="o", color="orange")
plt.title("Total Sales over time")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```
